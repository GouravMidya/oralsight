```python
from ultralytics import YOLO
import supervision as sv
from PIL import Image
import cv2
import numpy as np
import os
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from segment_anything import sam_model_registry, SamPredictor

# Load the YOLO model
model = YOLO('/content/best.pt')

# Load the test image
image_path = '/content/test.jpg'  # path to your test image
image = Image.open(image_path)
image_np = np.array(image)  # Convert to numpy array for further processing

# Perform inference on the image with YOLO
results = model(source=image_path, conf=0.25)[0]
print("YOLO inference completed.")

# Convert results to Supervision detections format
detections = sv.Detections.from_ultralytics(results)

# Load the SAM model and its predictor
sam_checkpoint = "/content/sam_vit_b.pth"  # Update with your SAM checkpoint path
sam = sam_model_registry["vit_b"](checkpoint=sam_checkpoint)
predictor = SamPredictor(sam)

# Set the image for SAM predictor (in RGB format)
predictor.set_image(image_np)

# Load the third model (dental radiography classification)
classification_model = tf.keras.models.load_model('/content/dental_radiography_model.h5')

# Class labels for classification model
class_labels = ['cavity', 'fillings', 'impacted tooth', 'implant', 'normal']

# Folder to save unsegmented and segmented parts
unsegmented_output_folder = '/content/teeth_collection/'
segmented_output_folder = '/content/segmented_teeth_collection/'
if not os.path.exists(unsegmented_output_folder):
    os.makedirs(unsegmented_output_folder)
if not os.path.exists(segmented_output_folder):
    os.makedirs(segmented_output_folder)

# Function to preprocess the image for classification
def preprocess_image_for_model(img):
    img = cv2.resize(img, (224, 224))  # Resize to match model input size
    img_array = np.expand_dims(img, axis=0)  # Expand dimensions to fit batch format
    img_array = img_array / 255.0  # Normalize the image
    return img_array

# Iterate over each detected bounding box
for idx, (x1, y1, x2, y2) in enumerate(detections.xyxy):  # YOLO bounding box format (x1, y1, x2, y2)

    # Crop the unsegmented tooth from the original image
    unsegmented_tooth = image_np[int(y1):int(y2), int(x1):int(x2)]

    # Save the unsegmented tooth
    unsegmented_output_path = os.path.join(unsegmented_output_folder, f'unsegmented_tooth_{idx}.png')
    cv2.imwrite(unsegmented_output_path, unsegmented_tooth)
    print(f'Unsegmented tooth saved at: {unsegmented_output_path}')

    # Convert bounding box to a format that SAM expects
    bbox = [x1, y1, x2, y2]
    input_box = np.array(bbox)

    # Perform segmentation using SAM
    masks, scores, logits = predictor.predict(
        point_coords=None,
        point_labels=None,
        box=input_box[None, :],  # SAM expects a batch of boxes
        multimask_output=False  # We are using a single mask
    )

    # Extract the first mask (since we requested a single mask)
    mask = masks[0]

    # Use the mask to segment the tooth from the original image
    segmented_tooth = cv2.bitwise_and(image_np, image_np, mask=mask.astype(np.uint8))

    # Crop the image to the bounding box (removing extra background)
    cropped_tooth = segmented_tooth[int(y1):int(y2), int(x1):int(x2)]

    # Preprocess the cropped tooth for the classification model
    preprocessed_tooth = preprocess_image_for_model(cropped_tooth)

    # Run the classification model on the preprocessed image
    predictions = classification_model.predict(preprocessed_tooth)

    # Get the predicted class and confidence score
    predicted_class = np.argmax(predictions, axis=1)[0]
    confidence_score = np.max(predictions) * 100  # Convert to percentage

    # Output the predicted class and confidence
    print(f"Detected tooth {idx}:")
    print(f"Predicted Class: {class_labels[predicted_class]}")
    print(f"Confidence Score: {confidence_score:.2f}%")

    # Optionally save the segmented tooth part for reference
    segmented_output_path = os.path.join(segmented_output_folder, f'segmented_tooth_{idx}.png')
    cv2.imwrite(segmented_output_path, preprocessed_tooth[0] * 255)  # Save the tooth after preprocessing
    print(f'Segmented tooth saved at: {segmented_output_path}')
```
